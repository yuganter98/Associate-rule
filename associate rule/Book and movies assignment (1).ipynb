# ACCOCIATE RULE ASSIGNMENT

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv('D:\\Assignment Excelr\\associate rule\\book.csv')
book
```

```python
book.head()
```

```python
df=pd.get_dummies(book)
df.head(10)
```

```python
df
```

```python
#MIN SUPPORT (apriori algo)
```

```python
frequent_itemsets=apriori(df,min_support=0.4,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets=apriori(df,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
#At 10% it takes all value and at 20% its take 7 terms and at 40% it take only two terms so it changes according to value we put in min support
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False).head(15)
```
